```python
import pandas as pd
do_num = eval(input('how much tasks for this month？'))
week_task_hrs = []
for i in range(1,do_num+1):
    do_name = input('No.'  +   str(i) +  ' task ')
    week1 = input('hours in week 1:')
    week2 = input('hours in week 2:')
    week3 = input('hours in week 3:')
    week4 = input('hours in week 4:')
    week_task_hrs.append({'task': do_name, 'week1' : week1,'week2':week2,'week3':week3,'week4':week4})

print(week_task_hrs)
df = pd.DataFrame(week_task_hrs)
df.to_csv('week_task_hrs.csv', mode='w',index=False)
df
```

```python
import pandas as pd
from matplotlib import pyplot as plt 

import random


def randomcolor():
    colorArr = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
    color = ""
    for i in range(6):
        color += colorArr[random.randint(0,14)]
    return "#"+color

df = pd.read_csv("week_task_hrs.csv") 
df
week_task_hrs= df.values.tolist()
print("task list:","\n",week_task_hrs)


task_name = []
hrs=[]

for i in week_task_hrs:
    task_name.append(i[0])
    hrs.append(i[1:5])

print(hrs)

task_name.append('week_mean')
print(task_name)

week_1=[]
week_2=[]
week_3=[]
week_4=[]
for i in week_task_hrs:
    week_1.append(i[1])
    week_2.append(i[2])
    week_3.append(i[3])
    week_4.append(i[4])
print(week_1)
print(week_2) 
print(week_3)
print(week_4)


mean_1=round(sum(week_1)/len(week_1),1)
mean_2=round(sum(week_2)/len(week_2),1)
mean_3=round(sum(week_3)/len(week_3),1)
mean_4=round(sum(week_4)/len(week_4),1)
print(mean_1)
print(mean_2)
print(mean_3)
print(mean_4)

week_mean=[mean_1,mean_2,mean_3,mean_4]
print(week_mean)


week_of_the_month = ['week1','week2','week3','week4']
for i in hrs:
   
    plt.plot(week_of_the_month,i,color=randomcolor(),marker='s')


plt.plot(week_of_the_month,week_mean,color=randomcolor(),linestyle='--',label='average')  
plt.xlabel("week of the month")
plt.ylabel("total_hrs_spent_per_week")
plt.legend(task_name)
plt.title('task week hrs')
plt.show()

```
